```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
s=['year','per_capita_income']
df=pd.read_csv('Exercise.csv',usecols=s)
df
```

```python
df.at[11,'year']=1981
df
```

```python
df=df.astype({'year':'int'})
df.per_capita_income.values
```

```python
plt.scatter(df.year,df.per_capita_income)
```

```python
x=df.drop('per_capita_income',axis='columns')
x
```

```python
y=df.per_capita_income
y
```

```python
reg= linear_model.LinearRegression()
```

```python
reg.fit(x,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
year_predic={'year':[2021,2022,2023,2024,2024]}
year_p=pd.DataFrame(year_predic)
year_p
```

```python
p_df=pd.DataFrame()
p_df['years']=year_p
```

```python
r_df=reg.predict(year_p)
```

```python
r_df
```

```python
p_df['Predicted Salary']=r_df
```

```python
p_df.head()
```

```python
p_df.to_csv("prediction.csv")
```
